# Handling Missing Values
Before proceeding to use the data we need to check for missing values to prevent any discrepencies. We can either fill the missing values with a value, for example the mean, median, etc. Or we can drop them.

```python
# Checking for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Checking for missing values
df.count()
```

By comparing the count of columns and their missing values, we decided to retain columns with missing values. Further inspection revealed that some columns are interrelated; specifically, if a binary column is 0, the corresponding column tends to be empty.
